Método de Bisección====================
El método de bisección es un algoritmo numérico utilizado para encontrar raíces de funciones continuas. Se basa en el teorema del valor intermedio, que establece que si una función cambia de signo en un intervalo cerrado, entonces debe tener al menos una raíz en ese intervalo.

El proceso de bisección implica dividir el intervalo en dos partes iguales y evaluar la función en el punto medio. Dependiendo del signo de la función en el punto medio, se selecciona el subintervalo que contiene la raíz y se repite el proceso hasta que se alcance una aproximación suficientemente precisa de la raíz.

````python
print('Evaluación de funciones en sympy')

import sympy as sp

x = sp.symbols('x')

func1 = (x**5 - 36*x)/(x**2+4)
sp.pprint(func1)

print('\nEvalúe a la función en un numero x0 usando subs')
print('El output se despliega con fracciones o números transcendetales porque sympy trabaja con algebra exacta')

x0 = 3
x1=func1.subs(x,x0)
sp.pprint(x1)
x2=func1.subs(x,x1)
sp.pprint(x2)
x3=func1.subs(x,x2)
sp.pprint(x3)


print('Use sp.float para desplegar estos números como decimales')
print('x1  ', sp.Float(x1))
print('x2', sp.Float(x2))
````

Uso del comando lambdify en SymPy====================
El comando `lambdify` en SymPy se utiliza para convertir expresiones simbólicas en funciones numéricas que pueden ser evaluadas de manera eficiente. Esto es especialmente útil cuando se desea evaluar una función simbólica en varios puntos sin tener que realizar la evaluación simbólica cada vez.
El comando `lambdify` toma una expresión simbólica y devuelve una función que puede ser llamada con valores numéricos. Esto permite aprovechar la velocidad de las evaluaciones numéricas mientras se mantiene la flexibilidad de trabajar con expresiones simbólicas.




````python
F2= (x**4 +3*x*5)/(x**3+5)
F2= sp.lambdify(x,F2)
x0 = 3

print(F2(x0))
````
