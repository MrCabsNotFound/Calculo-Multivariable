3. Uso de Matplotlib para graficar funciones
```python

````python
print('A. Generar vectores con np arrange y no.linspace')
x1 = np.arange(-4,5.01,1.5) #1.5 es el salto entre cada número
print('np.arrange',x1)
x2 = np.linspace(-4,5,9) #7 es el número total de puntos entre a y b (incluyéndolos)
print('np.linspace',x2)
print('Valores default para el tercer input de arange salto = 1 y linspace, puntos = 50')
#Si no colocan el tercer input, np.arrange tiene un salto default de 1
print(np.arange(-1,6))
print(np.linspace(-4,6))
````

````python
import matplotlib.pyplot as plt
t1 = np.linspace(-5,5,6) #Variable independiente
y1 = np.atan(t1) #Variable dependiente

plt.plot(t1,y1)
plt.title('Gráfica de tangente inverso con baja resolución')
plt.xlabel('tiempo (min.)')
plt.ylabel('altura (pies.)')
plt.grid() #Agregue una cuadrícula
plt.ylim(-1.7,1.7)
plt.xlim(-5,5)
plt.figure() #Abre una nueva ventana para las gráficas posteriores
#Utilice más puntos para mejorar la resolución de la gráfica
t2 = np.linspace(-5,5,200)
plt.plot(t2,np.atan(t2))
````

````python
t1 = np.linspace(-6,6,1000)
y1 = np.tanh(t1)
y2 = np.log(abs(t1))

#plt.plot(t1,y1,t1,y2)
plt.plot(t1,y1, 'r-')
plt.plot(t1,y2, 'k--')
plt.ylim([-3,3])
plt.legend(['tanh(t)', 'log|t|'],loc='lower right')
#Sombree la región entre las dos curvas usando plt.fill
plt.fill_between(t1,y1,y2, where=(y2 > y1), alpha=0.6)

````

6. Gráficas polares con plt.polar

````python
theta = np.linspace(0,2*np.pi,1000)
r1 = 4*np.cos(3*theta)
r2 = 4*np.sin(4*theta)

plt.polar(theta,r1)
plt.polar(theta,r2)
plt.title('Comparativo curvas polares')
plt.legend(['Rosa 3 pétalos', 'Rosa 4 pétalos'])
plt.fill_between(theta,r1,r2,where=(r1>r2))
````
