Método de Bisección====================
El método de bisección es un algoritmo numérico utilizado para encontrar raíces de funciones continuas. Se basa en el teorema del valor intermedio, que establece que si una función cambia de signo en un intervalo cerrado, entonces debe tener al menos una raíz en ese intervalo.

El proceso de bisección implica dividir el intervalo en dos partes iguales y evaluar la función en el punto medio. Dependiendo del signo de la función en el punto medio, se selecciona el subintervalo que contiene la raíz y se repite el proceso hasta que se alcance una aproximación suficientemente precisa de la raíz.

Metodo Biseccion, ciclo for

````python
import sympy as sp
from tabulate import tabulate

x = sp.symbols('x')

F3 = x**6 - 45*x    #Función
a = 2               #Extremos del intervalo
b = 3
Pasos = 6  
Precision = 0.5e-4  #Cuatro Decimales Correctos    

#Adentro del método de Bisección
F3 = sp.lambdify(x,F3)
#Tabla en Blanco
Tabla =[]   #Inicializaciones de las variables
ErrorAprox = 1  
Pasos = 1

#print('a         b              c            fa                      fc             fb')

while ErrorAprox>Precision:

    Pasos +=1
    ErrorAprox = 0.5*(b-a)
    c = 0.5*(a+b)   #Punto Medio
    Tabla.append([Pasos,a,c,b,F3(a), F3(c), F3(b),ErrorAprox]) #Impresión de la tabla

    #Actualiza el intervalo utilizando un condicional if
    if F3(a)*F3(c)<0:
        b = c  #Actualice el extremo derecho
    else:
        a = c #Actualice el extremo izquierdo

print(tabulate(Tabla,headers=['Pasos','a','c','b','f(a)','f(b)','f(c)','Error Aproximado'],
               numalign='center',tablefmt='fancy_grid',
               floatfmt=['3d','13.10f','13.10f','13.10f','6.2e','6.2e','6.2e','6.3e']))

    #print('Nuevo intervalo:  [', a, b,']')
````

Definiciones para el método de bisección


````python
def MetodoBiseccion(Funcion,xizq=-4,xder=4,Precision=1e-6):



    #Adentro del método de Bisección
    Funcion = sp.lambdify(x,Funcion)   #Facilita Evalución Numérica
    #Tabla en Blanco
    Tabla =[]   #Inicializaciones de las variables
    ErrorAprox = 1  
    Pasos = 0

    while ErrorAprox>Precision:

        Pasos +=1
        ErrorAprox = 0.5*(xder-xizq)
        medio = 0.5*(xizq+xder)   #Punto Medio
        Tabla.append([Pasos,xizq,medio,xder,Funcion(xizq), Funcion(medio), Funcion(xder),ErrorAprox]) #Impresión de la tabla

        #Actualiza el intervalo utilizando un condicional if
        if Funcion(xizq)*Funcion(medio)<0:
            xder = medio  #Actualice el extremo derecho
        else:
            xizq = medio #Actualice el extremo izquierdo
    print('Tabla de Resultados Método de Bisección')
    print(tabulate(Tabla,headers=['Pasos','Izquierdo a','Medio c','Derecho b','f(a)','f(b)','f(c)','Error Aproximado'],
                numalign='center',tablefmt='fancy_grid',
                floatfmt=['3d','13.10f','13.10f','13.10f','6.2e','6.2e','6.2e','6.3e']))
    

def MetodoNewton(Funcion,xest=-4,Precision=1e-6):


    Derivada = sp.diff(Funcion,x)
    #Adentro del método de Bisección
    Funcion = sp.lambdify(x,Funcion)   #Facilita Evalución Numérica
    Derivada = sp.lambdify(x,Derivada)   #Facilita Evalución Numérica

    #Tabla en Blanco
    Tabla =[]   #Inicializaciones de las variables
    ErrorAprox = 1  
    Pasos = 0

    while ErrorAprox>Precision:

        Pasos +=1
        nuevo = xest-Funcion(xest)/Derivada(xest)   #Estimación de la Raíz
        ErrorAprox = abs(nuevo-xest)
        xest = nuevo  #Actualice la estimación inicial
        Tabla.append([Pasos,nuevo,Funcion(nuevo),ErrorAprox]) #Impresión de la tabla

    print('Tabla de Resultados Método de Newton')
    print(tabulate(Tabla,headers=['Pasos','Raíz Estimada','f(nuevo)','Error Aproximado'],
                numalign='center',tablefmt='fancy_grid',
                floatfmt=['3d','19.16f','6.2e','6.2e']))

````
