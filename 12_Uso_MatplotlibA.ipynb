````python
print('Una Tabla Rudimentaria \n')

import numpy as np

x= np.linspace(-3,5,9)
y = x**3-3*x**2-6*x

for indice in range(0,9):
    print(x[indice],y[indice])

print('\n Mejoras a la tabla con un encabezado, texto formateado, range dinámico')

print('Distancia (m)    Voltaje (V)')
print('-------------------------------')

for indice in range(0,len(x)):
    print('   {:5.1f}          {:7.3f}'. format(x[indice],y[indice]))


````

Tablas automatizadas con tabulate

````python
from tabulate import tabulate
tabla =list(zip(x,y))
print('Tabla can tabulate sencilla')
print(tabulate(tabla))
#Tabulate tiene funciones adicionales para agregar encabezados, alinear texto, formato de números, etc.
print('Una tabla con encabezados')
tabla2= tabulate(tabla,headers=['Distancia(m)','Voltaje (V)'])
print(tabla2)
#print('Una tabla con formato de número, alineación y bordes')
tabla3= tabulate(tabla,headers=['Distancia(m)','Voltaje (V)'],floatfmt=['7.1f','8.2f'],tablefmt='fancy_grid',numalign='center')
print(tabla3)
````

3.  Aplicación:  Trayectoria de un proyectil que tiene los siguientes parametros

velocidad inicial: Vo

ángulo de lanzamiento  $\theta$

altura inicial: $y_0$

Aceleración gravitacional: g

Ecuaciones para la altura $y$ y la ditancia horizontal $x$ en función del tiempo $t$:

\begin{align*}
x &= V_o \cos(\theta) t \\
y &= y_0 +V_o \sin(\theta) t - \frac{1}{2} g t^2
\end{align*}

````python
import matplotlib.pyplot as plt

print('Parametros del modelo')

Veloc_Inic = 35 #m/s
Angulo_Inic = 60 #grados
Altura_Inic = 10 #m
Aceleracion = 9.81 #m/s^2
Tiempo_Inic = 0 #s
Tiempo_Fin  = 6 #s
Num_Puntos  = 14

#Variables que se generan proporcionados los parametros
Tiempo = np.linspace(Tiempo_Inic,Tiempo_Fin,Num_Puntos)
AnguloR = np.radians(Angulo_Inic)
Distancia_x = Veloc_Inic*Tiempo*np.cos(AnguloR)
Altura_y = Veloc_Inic*Tiempo*np.sin(AnguloR) - 0.5*Aceleracion*Tiempo**2

#Imprima la tabla con los valores de t,x,y

Tabla = list(zip(Tiempo,Distancia_x,Altura_y))
Tabla1 = tabulate(Tabla,headers=['Tiempo (s)','Distancia (m)', 'Altura (m)'],tablefmt='rounded_grid',numalign='center')
print(Tabla1)

plt.plot(Distancia_x,Altura_y)
plt.title('Trayectoria Proyectil para un ángulo de {:3.1f} y velocidad inicial de {:4.2f}'. format(Angulo_Inic,Veloc_Inic))
plt.xlabel('Distancia (m.)')
plt.ylabel('Altura (m.)')
plt.grid()
````

def trayectoria_proyectil(Vo, theta, y0, g, t):
    x = Vo * np.cos(theta) * t
    y = y0 + Vo * np.sin(theta) * t - 0.5 * g * t**2
    return x, y

La función también va a proporcionar una tabla de valores y una gráfica de la trayectoria del proyectil.

````python
def trayectoria_proyectil(Veloc_Inic,Angulo_Inic,Altura_Inic,Aceleracion=9.81,Tiempo_Inic=0,Tiempo_Fin=10,Num_Puntos=30,VerTabla='N'):

    """""
    Trayectoria de un proyectil en distancia x y en altura y

    x = vo*t cos(tetha)

    y = v*t sin(tetha) -0.5*g*t^2 + y0

    EL usuario tiene que proporcionar 

    Veloc_inicial

    Altura_inic

    Angulo (en grados)

    Los inputs opcionales son:
    
    Aceleracion = 9.81 m/s^2
    
    Tiempo_Inic = 0
    
    Tiempo_Fin  = 10
    
    Num_Puntos  = 30
    
    VerTabla    = 'N' (no se despliega)
    """""

    #Variables que se generan proporcionados los parametros
    Tiempo = np.linspace(Tiempo_Inic,Tiempo_Fin,Num_Puntos)
    AnguloR = np.radians(Angulo_Inic)
    Distancia_x = Veloc_Inic*Tiempo*np.cos(AnguloR)
    Altura_y = Veloc_Inic*Tiempo*np.sin(AnguloR) - 0.5*Aceleracion*Tiempo**2

    #Imprima la tabla con los valores de t,x,y

    if VerTabla in ['Si','Y','S','Yes']: 
    #or VerTabla == 'Si' or VerTabla == 'Y' or VerTabla == 'Si' :
        Tabla = list(zip(Tiempo,Distancia_x,Altura_y))
        Tabla1 = tabulate(Tabla,headers=['Tiempo (s)','Distancia (m)', 'Altura (m)'],tablefmt='rounded_grid',numalign='center')
        print(Tabla1)

    plt.plot(Distancia_x,Altura_y)
    plt.title('Trayectoria Proyectil para un ángulo de {:3.1f} y velocidad inicial de {:4.2f}'. format(Angulo_Inic,Veloc_Inic))
    plt.xlabel('Distancia (m.)')
    plt.ylabel('Altura (m.)')
    plt.grid()
    plt.figure()

    return Tiempo,Distancia_x,Altura_y

print('Ejemplos de Prueba')

#Caso con todos los inputs proporcionados
t1,x1,y1 = trayectoria_proyectil(100,120,-30,6,0,30,30,'Y')
#Caso con pocos inputs y usando valores default
t2,x2,y2 = trayectoria_proyectil()

    
````
