**Uso de Tablas**
ED de resistencia al aire lineal
$$\frac{dv}{dt} = -k \left(v - \frac{mg}{k} \right), \qquad      v(0)=v_0$$

La solución de la ED es:
$$v(t) = \frac{mg}{k} + \left( v_0 - \frac{mg}{k} \right) e^{-kt}$$

```python
import numpy as np
import matplotlib.pyplot as plt

#Parametros del Modelo
m = 2.0 # Masa del objeto (kg)
g = 9.81 # Aceleración gravitacional (m/s^2)
k = 0.2  #Constante de resistencia al aire (1/s)
V0 = 1   #Velocidad inicial  (m/s)
Vterm = m*g/k    #Velocidad Terminal (m/s)

tiempo = np.linspace(0,25,26) # Tiempo de 0 a 20 segundos, 21 puntos
Velocidad = Vterm + (V0-Vterm)*np.exp(-k*tiempo)
```

**Construcción de una tabla de valores con un ciclo for**

```python

#Mejoras en el despliegue de la tabla
print('Velocidad Terminal (m/s):  {:8.4f} . \n'. format(Vterm))
print("\nTiempo (s) | Velocidad (m/s)")
for i in range(0,len(tiempo),2):  # Muestra cada segundo valor
    print(f"{tiempo[i]:10.2f} | {Velocidad[i]:10.4f}")
```

Construcción de una tabla con tabulate

```python
from tabulate import tabulate

print(tabulate(zip(tiempo, Velocidad), headers=["Tiempo (s)", "Velocidad (m/s)"], tablefmt='rounded_grid',floatfmt=('.2f','.4f')))  # Tabla con formato de cuadrícula
#print(tabulate(zip(tiempo, Velocidad), headers=["Tiempo (s)", "Velocidad (m/s)"], tablefmt='latex'))  # Tabla con formato latex
      
```

Definiciones, Funciones que aceptan argumentos

```python
def ResistenciaAire(m,g,k,V0=0,TN=40,puntos=21,Graficar='Si',Tabla='No'):
    
    """
    Encuentra la velocidad de un objeto sujeto a resistencia de aire lineal
    Imprime la tabla de resultados
    
    Los parámetros inputs de la función son:
    
     m = masa del objeto (kg.)
     g = Aceleración gravitacional (m/s^2)
     k = constante de resistencia al aire (1/s)
     V0 = velocidad inicial  (m/s)  (valor default en reposo)
     TN = tiempo final (s)  (valor default 40 segundos)
     puntos = número de datos que se desean desplegar (valor default 41 puntos)
     Graficar  despliega la gráfica (default, si)
     Tabla     despliega la tabla   (default, no)

    """
    
    tiempo = np.linspace(0,TN,puntos)
    Vterm  = m*g/k
    Velocidad = Vterm + (V0-Vterm)*np.exp(-k*tiempo)  #Calcula la velocidad en cualquier momento

    if Tabla == 'Si' or Tabla =='Yes' or Tabla == 1:
        print('Velocidad Terminal del Objeto: {:6.4f} m/s. \n'. format(Vterm))
        print('Tabla de Resultados para m = {:4.1f}, g = {:4.2f}, k = {:1.3f}'. format(m,g,k))
        print(tabulate(zip(tiempo, Velocidad), headers=["Tiempo (s)", "Velocidad (m/s)"], tablefmt='rounded_grid',floatfmt=('.2f','.4f')))  # Tabla con formato de cuadrícula

    if Graficar == 'SI' or Graficar == 'Si':
        plt.plot(tiempo,Velocidad)
        plt.plot(tiempo, Vterm*np.ones_like(tiempo), 'r--', label='Velocidad Terminal')
        plt.title('Velocidad de un objeto con resistencia al aire para m = {:4.1f}, g = {:4.2f}, k = {:1.3f}'. format(m,g,k))
        plt.xlabel('Tiempo (s)')
        plt.ylabel('Velocidad (m/s)')
        plt.grid()
```
