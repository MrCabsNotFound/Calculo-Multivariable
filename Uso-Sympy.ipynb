import sympy as sp

```python
import sympy as sp
x,y,z = sp.symbols("x,y,z") #Variables Simbólicas

w = 9 #No simbólica

I1 = sp.cos(x**2+y**2) #Función con dos variables independientes

print("Utilice print o sp.pprint para desplegar la función")
print(I1)
print("Impresión más bonita ")
sp.pprint(I1)
```

```python
DIx = sp.diff(I1,x)
print("Derivada parcial de I1 respecto a x")
print(DIx)
print("Derivada parcial de I1 respecto a y")
print(sp.diff(I1,y))

print("\nLas derivadas de orden superior también se encuentran con sp.diff(f,(x,n))")
DIxx = sp.diff(I1,(x,2))
print("2da derivada de I1 respecto a x x")
print(DIxx)
print("3ra derivada de I2 respecto a y y y")
print(sp.diff(I1,(x,3)))

print("\nLa derivada parcial cruzada Ixy requiere dos veces el uso de diff")
Ixy = sp.diff(sp.diff(I1,x),y)
Iyx = sp.diff(sp.diff(I1,y),x)
print("Las derivadas parciales cruzadas son iguales")
print(Ixy)
print(Iyx)
```

Integración con sp.integrate()

Integrate = cos(x**2+y**2)

```python
F1 = sp.integrate(I1,x)
print(F1)

F2 = 2*x*y**3*sp.sin(x**2+y**2)
print("F2 si se puede integrar por sustitución")
print(sp.integrate(F2,x))
print("F2 si se puede integrar respecto a y usando IPP y sustitución")
print(sp.integrate(F2,y))

```

Integrales Dobles, porque sp.integrate() también tiene límites de integración

```python
I2 = 42*x/(sp.sqrt(1+y**7))

print("\nIntegrar por fuerza bruta respecto a y")
print(sp.integrate(I2,y))

print("\nCambie el orden de integración e integrar de primero respecto a x")
print(sp.integrate(I2,x))
print("Evalúe la integral indefinida en 0<x<y^3")
I2x = sp.integrate(I2,(x,0,y**3))
print(I2x)

print("\nAhora realice la integración en y")
sp.pprint(sp.integrate(I2x,y))
print("Evalúe en los límites de integración 0 < x < 3^(1/7")
print(sp.integrate(I2x,(y,0,3**(1/7))))
```
